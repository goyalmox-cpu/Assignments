* Use NumPy to create at least three pairs of vectors.
* Calculate and print their dot products.
* Experiment with different values and observe the results.

```python
import numpy as np
```

```python
#pair 1
a = np.array([2, 3])
b = np.array([4, 5])
dot_prod = np.dot(a,b)
print(dot_prod)
```

```python
# pair 2
c = np.array([1, 2, 3])
d = np.array([4, 5, 6])
dot_prod = np.dot(c,d)
print(dot_prod)
```

```python
# Pair 3
e = np.array([7, 8])
f = np.array([2, 1])
dot_prod = np.dot(e,f)
print(dot_prod)
```

# Experiment 1: Orthogonal vectors


```python
g = np.array([1, 0])
h = np.array([0, 1])
dot_prod = np.dot(g, h)
print(dot_prod)
```

# Experiment 2: Same direction vectors


```python
i = np.array([5, 5])
j = np.array([5, 5])
dot_prod = np.dot(i,j)
print(dot_prod)
```

## Experiment 3: Opposite direction vectors


```python
k = np.array([1, 2])
l = np.array([-1, -2])
dot_prod = np.dot(k,l)
print(dot_prod)
```
