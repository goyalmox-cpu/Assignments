# Eigenvalues and Eigenvectors


```python
C = np.array([[2,0],[0,3]])
```

```python
values1, vectors1 = np.linalg.eig(C)
```

```python
print(C)
```

```python
print("Eigenvalues:")
print(values1)
print(" ")
print("Eigenvectors:")
print(vectors1)
```

```python
D = np.array([[4,1],[2,3]])
```

```python
values2, vectors2 = np.linalg.eig(D)
```

```python
print(D)
```

```python
print("Eigenvalues:")
print(values2)
print(" ")
print("Eigenvectors:")
print(vectors2)
```
