## Features vs Target Variable
* Features (X): input variables for making pred.
* Target Variable (y): output whivh model is pred.
* eg: in house pred, data set: features are bedrooms and target is house price.

# ML WORKFLOW
Data Collection->
Data Cleaning->
Feature Selection->
Model Training->
Evaluation->
Prediction

1. Data collection: gathering data for training for ML model.
--> sources:
* CSV/Excel files
Databases
APIs
Sensors and IoT devices
Surveys
Kaggle datasets
Web scraping

2. Data cleaning:
Missing values
Duplicate records
Incorrect values
Typing mistakes
Outliers
Inconsistent formats

3. feature selections: Selecting only the useful input variables (features) that help predict the target.Not every column is useful.

4. Model Training: The cleaned data is given to a machine learning algorithm, which learns patterns and relationships between features and the target.

5. Evaluation: After training, we test how well the model performs on unseen data.
"Can the model make accurate predictions on new data?"

6. Prediction: Once the model performs well, it is used to make predictions on new, unseen data.

## Mini Exploration


Why is data called the "fuel" of Machine Learning?

Machine learning models learn patterns from data. Just as a car cannot run without fuel, an ML model cannot learn without data. Better-quality and more diverse data usually leads to better predictions.

What happens if poor-quality data is used?

Poor-quality data can contain errors, missing values, or incorrect labels. Models trained on such data often make inaccurate predictions and may fail to generalize to new data.

Why is the Iris dataset popular?

The Iris dataset is small, clean, and easy to understand. It has no missing values.
